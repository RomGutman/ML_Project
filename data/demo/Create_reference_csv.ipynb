```python
import pandas as pd
import os
import xml.etree.ElementTree as ET
```

```python
path = os.path.abspath(os.curdir)
```

```python
tree = ET.parse('SintecMedia.rpt_refxml.date_2015-01-04.2016-11-21.xml')
root = tree.getroot()
```

```python
elm_list = {}
mso = {}
for idx, child in enumerate(tree.getiterator()):
    if child.tag == 'mso-table':
        mso = {'mso_code': child.attrib['mso'], 'mso_name': child.attrib['mso-name']}
    if child.tag == 'mapping':
        elm_list[idx] = child.attrib
        elm_list[idx].update(mso)
```

```python
reference = pd.DataFrame(elm_list).T
reference['household-id'] = reference['household-id'].astype(int)
```

```python
sample = pd.DataFrame()
for i in range(14):
    i=i+1
#     sample1 = pd.read_csv('Samples'+str(i)+'.csv')
    sample = sample.append(pd.read_csv('Samples'+str(i)+'.csv'), ignore_index=True)
sample.shape
```

```python
demograph = pd.merge(sample, reference, how='inner', left_on=['Household ID'], right_on=['household-id'])
```

```python
demograph.to_csv('demographic.csv')
demograph.head(10)
```
